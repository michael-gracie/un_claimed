```python
import pandas as pd
```

```python
df = pd.read_csv('data/heart_prop.csv')
```

```python
df['cum_amt'] = df.sort_values(by='Amount',ascending = False)['Amount'].cumsum()
df['amt'] = df['Amount'].sum()
df['amt_per'] = df['cum_amt']/df['amt']
```

```python
df[df['amt_per'].lt(0.8)].sort_values(by='Amount')
```
