## Getting list of all charities

Checked the T&C's no word on scraping https://www.canada.ca/en/transparency/terms.html?request_locale=en

```python
charity_df = pd.DataFrame()
for i in range(0,858):
    print(f'Page {i+1}')
    df_list = pd.read_html(f'https://apps.cra-arc.gc.ca/ebci/hacc/srch/pub/bscSrch?dsrdPg={i+1}&q.stts=0007&q.ordrClmn=NAME&q.ordrRnk=ASC')
    charity_df = charity_df.append(df_list[0], ignore_index=True)
    sleep(random.randint(3,5))
```

```python
charity_df.to_csv('data/can_charity.csv', index = False)
```

```python
master_df = pd.DataFrame()
```

```python
keywords = [
    'heart'
    ,'charity'
    ,'salvation'
    ,'foundation'
    ,'hospital'
    ,'health'
    ,'cancer'
]
```

```python
for keyword in keywords:
    df = get_data.pull_properties(keyword)
    master_df = master_df.append(df, ignore_index=True)
```

```python
master_df['Amount'] = master_df['Amount'].astype(float)
```

```python
master_df[master_df['Payee'].str.contains('SALVATION ARMY')].sort_values(by = 'Amount', ascending = False)['Amount'].sum()
```
